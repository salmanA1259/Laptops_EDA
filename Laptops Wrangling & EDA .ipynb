# Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
laptops = pd.read_csv("laptops.csv", encoding="latin-1") # windows-1251, latin-1
```

```python
laptops.columns = \
laptops.columns.str.strip().str.replace("(","").str.replace(")","").str.lower().str.replace(" ","_")\
.str.replace("operating_system","os")
```

```python
laptops.columns
```

```python
laptops['screen_size'] = laptops['screen_size'].str.strip('"').astype(float)
```

```python
laptops['screen_size'].mean()
```

```python
#laptops = laptops.rename({"screen_size":"screen_size_inches"}, axis = 1)
laptops.rename({"screen_size":"screen_size_inches"}, axis = 1, inplace = True)
```

```python
laptops
```

```python
"Intel Iris Plus Graphics 640".split()[0]
```

```python
laptops["gpu_manufacturer"] = \
laptops["gpu"].str.split(n=1, expand = True)[0]
```

```python
laptops["gpu_manufacturer"].value_counts()
```

```python
# task CPU
```

```python
laptops["cpu_type"] = \
laptops["cpu"].str.split(n=1, expand = True)[0]
```

```python
laptops["cpu_type"].value_counts()
```

```python
laptops["os"].unique()
```

```python
#laptops["os"].replace("Mac OS","macOS")
```

```python
laptops.loc[laptops["os"] == "Mac OS", "os"] = "macOS"
```

```python
laptops["os"].unique()
```

```python
laptops.isnull().sum()
```

```python
laptops.loc[laptops["os_version"].isnull(),"os"].value_counts()
```

```python
laptops.loc[laptops["os_version"].isnull(),"os_version"] = "Version Unknown"
```

```python
laptops.info()
```

```python
laptops["os_version"].unique()
```

```python
laptops['cpu'] = laptops['cpu'].apply(lambda x: ' '.join(x.split()[:3]) if x.startswith('Intel') else x)
laptops['cpu'] = laptops['cpu'].apply(lambda x: ' '.join(x.split()[:2]) if x.startswith('AMD') else x)
laptops['cpu'] = laptops['cpu'].apply(lambda x: ' '.join(x.split()[:2]) if x.startswith('Intel Atom') else x)
laptops['cpu'] = laptops['cpu'].apply(lambda x: ' '.join(x.split()[:2]) if x.startswith('Intel Xeon') else x)
laptops['cpu'] = laptops['cpu'].apply(lambda x: ' '.join(x.split()[:2]) if x.startswith('Samsung') else x)
```

```python
laptops["price_euros"] = laptops["price_euros"].str.replace(",", ".").astype(float)
laptops.info()
```

```python
laptops.to_csv("laptops_cleaned.csv", index = False)
```

# Number of Model released by manufacturers

```python
plt.figure(figsize=(15, 8))

plot = sns.countplot(data=laptops, y='manufacturer')

manufacturers = laptops['manufacturer'].unique()
manufacturer_counts = laptops["manufacturer"].value_counts().reindex(manufacturers)

for i, manufacturer in enumerate(manufacturer_counts):
    if (manufacturer > 200):
        plot.patches[i].set_facecolor('Green') 
    else:
        plot.patches[i].set_facecolor('LightGrey')

for i, value in enumerate(manufacturer_counts):
    if (value > 200):
        plt.text(value, i, str(value), ha="left", va='center', fontsize = 10, \
            fontweight= "bold", color = "Green")
    else:
        plt.text(value, i, str(value), ha="left", va='center', fontsize = 10, \
            fontweight= "bold", color = "grey")
    
plot.set_facecolor('none')
plt.yticks(rotation=0, color="Green")
plot.axes.xaxis.set_visible(False)
sns.despine(left=True, bottom= True)
plt.title("2017: The Sizzling Showdown of Laptop Royalty - HP, Lenovo, and Dell Take Center Stage!",\
    fontsize=15, fontweight='normal', color='DarkGreen', pad= 30, loc= "left")
plt.xlabel('')
plt.ylabel('')

#plt.show()
plt.savefig ("123.jpeg")
```

# Category Distribution Countplot

```python
cat = laptops['category'].value_counts()
category_counts = cat.sort_values(ascending=False).index


plt.figure(figsize=(8, 3))

plot = sns.countplot(data=laptops, y='category', order= category_counts)

for i, value in enumerate(category_counts):
    if (value == "Notebook"):
        plot.patches[i].set_facecolor('Green') 
    else:
        plot.patches[i].set_facecolor('lightgrey')

for i, value in enumerate(cat):
    if (value > 500):
        plt.text(value, i, str(value), ha="left", va='center', fontsize = 10, fontweight= "bold", color = "Green")
    else:
        plt.text(value, i, str(value), ha="left", va='center', fontsize = 10, fontweight= "bold", color = "grey")

plot.set_facecolor('#f7f7f7')
plt.xlabel('')
plt.ylabel('')
plt.yticks(rotation=0, color= "Green")
plot.axes.xaxis.set_visible(False)
sns.despine(left=True, bottom= True)
plt.title("Category Distribution", fontsize=15, fontweight='normal', color='DarkGreen', pad= 30, loc= "center")

plt.tight_layout()
# plt.show()
plt.savefig ("Category.jpg")
```

# OS Distribution Pie Chart

```python
# Count the occurrences of each OS
Total_os = pd.Series(laptops["os"]).value_counts()

# removing the values less than 10
Total_os = Total_os[Total_os>10]
Total_os
```

```python
# Create a pie chart
plt.figure(figsize=(6, 5))

colors=['#6bcb6b', '#8cdd8c', '#aee9ae', '#4daf4a', '#2ca02c']

plt.pie(Total_os, labels=Total_os.index, autopct='%1.1f%%', colors=['#6bcb6b', '#8cdd8c', '#aee9ae', '#4daf4a', '#2ca02c'])
plt.title('OS Distribution', fontsize=16, fontweight='normal', color='DarkGreen', pad= 30, loc= "center")

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.tight_layout()
# plt.show()
plt.savefig ("OS.jpg")
```

# Price vs Category Stripplot

```python
plt.figure(figsize=(15, 6))

snip = sns.stripplot(data=laptops, x="manufacturer", y="price_euros",hue="category", palette="Greens", dodge=True)
plt.xlabel("")
plt.ylabel("Price(Euros)", fontsize = 12, color= "Green")
plt.title('Price vs Category', fontsize=16, fontweight='normal', color='DarkGreen', pad= 30, loc= "center")


# Customize the appearance
plt.grid(False)
plt.yticks(range(0, 7000, 1000), fontsize=10, rotation= 0, color= "Green")
plt.xticks(fontsize=12, rotation = 45, color= "Green")
plt.gca().spines['top'].set_visible(0)
plt.gca().spines['right'].set_visible(0)
plt.gca().spines['bottom'].set_visible(0)
plt.gca().spines['left'].set_visible(0)

# Add a background color
plt.gca().set_facecolor('#f7f7f7')

plt.tight_layout()
# plt.show()
plt.savefig ("Price Category relation.jpg")
```
